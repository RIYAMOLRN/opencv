```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
image=cv2.imread("/content/inception.jpg")

(h,w)=image.shape[:2]
center=(w//2,h//2)#floor division for integer
angle=50
scale=1.0
rotation_matrix=cv2.getRotationMatrix2D(center,angle,scale)
rotated=cv2.warpAffine(image,rotation_matrix,(w,h))
plt.imshow(cv2.cvtColor(rotated,cv2.COLOR_BGR2RGB))
plt.title("ROTATED IMAGE (50)")
```

```python
flipped_horizontal=cv2.flip(image,1)

plt.subplot(2,2,1)#row column image
plt.imshow(cv2.cvtColor(flipped_horizontal,cv2.COLOR_BGR2RGB))
plt.title("FLIPPED HORIZONTAL")
```

```python
import numpy as np
tx,ty=30,-100#shift in x and y direction
translation_matrix=np.float32([[1,0,tx],[0,1,ty]])
translated=cv2.warpAffine(image,translation_matrix,(w,h))
plt.imshow(cv2.cvtColor(translated,cv2.COLOR_BGR2RGB))
plt.title("TRANSLATED IMAGE")
plt.show()
```
